```python
## 缺失值处理
### 8- 计算缺失值 总计
df = pd.read_csv(file.joinpath('TOP250.txt'), encoding='gbk', sep='\t')
df.isna().sum().sum()

### 9 - 计算缺失值｜分列
df.isnull().sum()

### 10 - 查看缺失值
df[df.isna()]

### 10 - 查看缺失值 为了后面更方便的处理缺失值，现在先看看全部缺失值所在的行
df.loc[df.isna().T.any(), :]

### 11- 高亮缺失值
(
    df.style
    .highlight_null(props="background-color:blue;color:yellow")
)

### 12- 删除缺失值
df.dropna(how='any')

### 13- 缺失值不全 | 整体填充
df.fillna('*')

### 14 - 缺失值补全｜向上填充
df.fillna(method='pad')

### 15 - 缺失值补全｜整体均值填充
df.评价人数.fillna(df.评价人数.mean())

### 16 - 缺失值补全｜上下均值填充 除了可以使用整列的均值进行填充，也可以使用缺失值位置的上下均值进行填充、

df.评价人数.fillna(df.评价人数.interpolate())

### 17 -缺失值补全｜匹配填充 除了利用均值填充，有时还需要根据另一列的值进行匹配填充

# -> 现在填充 “语言” 列的缺失值，要求根据 “国家/地区” 列的值进行填充

# 例如 《海上钢琴师》国家/地区为 意大利，根据其他意大利国家对应的语言来看，应填充为 意大利语

df.groupby('国家/地区').语言.bfill()

### 18 - 查找重复值 将全部重复值所在的行筛选出来
df[df.duplicated()]

### 19 -查找重复值｜指定 上面是所有列完全重复的情况，但有时我们只需要根据某列查找重复值
df[df.duplicated(['片名'])]

### 20 - 删除重复值
df = df.drop_duplicates()

### 21 - 删除重复值｜指定
df.drop_duplicates(keep='last')
```
