```python
## 2-3 基于 style 个性化设置
'''
上面基于 option 的 pandas 相关设置是全局配置，一次设置会在关闭notebook前一直有效

但相关常用的设置并不多，不能满足更多的个性化需求。

幸运的是在 pandas 中提供 Styler 对象让我们进一步个性化展示数据。

本节我就将一些常用的基于 style 个性化设置整理为习题模式方便大家学习、巩固。

注意：基于 style 个性化设置同样不会修改数据，所有 data.style.xxxx 输出的数据均是一次性的（可以复用、导出），

因此你应该在合适的时间选择使用该方法。
'''
data = pd.read_csv(file.joinpath("某招聘网站数据.csv"), usecols=[
                   'positionName', 'createTime', 'salary', 'subwayline', 'matchScore'], nrows=20, parse_dates=['createTime'])
data
# data.style.hide_index() 已经遗弃
data.style.hide(axis=0)
data

## 13 调整精度
pd.set_option("display.precision", 2)
data

## 高亮缺失值
data.style.highlight_null(props='color:white; font-weight:bold; background-color:darkblue;')

## 高亮最大值

data.style.highlight_max(subset='salary', props='color:white; font-weight:bold; background-color:darkblue;')

(
    data.style
    .highlight_max(subset='salary', props='color:white; font-weight:bold; background-color:red;')
    .highlight_min(subset='matchScore', props='color:white; font-weight:bold; background-color:orange;')
    # .highlight_between(left=left, right=, props='color:white; font-weight:bold; background-color:red;')
    .highlight_null(props='color:white; font-weight:bold; background-color:darkblue;')
)
## 同时高亮 最大值和最小值
(
    data
    .style
    .highlight_max(color='#F77802')
    .highlight_min(color='#26BE49')
)

(
    data
    .style
    .highlight_between(subset='salary', left=3000, right=30000, props="color:orange;background-color:blue; font-size:15;")
)

## 渐变显示列

## 修改字体颜色
(data
.style
.set_properties(
    subset=['salary'], **{'color': 'red'}))

# 22 修改背景颜色、对齐方式、字体大小
# 将整个 dataframe 进行如下设置：

# 居中
# 背景色修改为 #F8F8FF
# 字体:13px

(data
.style
.set_properties(**{'background-color': 'darkblue','text-align':'center', 'font-size': '13px'})
.set_properties(
    subset=['salary'], **{'color': 'pink'}))

(data
.style
.set_properties(**{'background-color': 'darkblue','text-align':'center', 'font-size': '13px','color':'white'})
.set_properties(
    subset=['salary'], **{'color': 'red'})).to_excel(file.joinpath('带有样式导出.xlsx'))


## 制作条形图
(data
.style
.set_properties(**{'background-color': 'orange'})
.bar(subset=['matchScore'],color='blue'))

## 带条件的样式
def my_style(val):
    color = 'red' if val > 30000 else 'black'
    return 'color: %s' % color


data.style.applymap(my_style, subset="salary")

## 指定格式化数据
(data
.style
.format("{0:,.2f}分", subset="matchScore")
.format("{""}元", subset="salary"))


## 格式化输出日期类型
data.style.format({"createTime": lambda t: t.strftime("%Y年%m月%d日")})
```
