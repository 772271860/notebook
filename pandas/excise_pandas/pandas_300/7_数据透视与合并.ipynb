```python
import numpy as np
from pathlib import Path
import pandas as pd

pd.set_option('display.max_rows', 15)

file = Path(r"C:\Users\Administrator\Desktop\pandas进阶修炼")
df = pd.read_csv(file.joinpath("某超市销售数据.csv"),thousands=',')
df.head()
```

```python
# 2 - 数据透视｜默认
# 制作各省「平均销售额」的数据透视表
df.pivot_table(index='省/自治区', values='销售额')
```

```python
# 3 - 数据透视｜指定方法
# 制作各省「销售总额」的数据透视表
df.pivot_table(index='省/自治区', values='销售额', aggfunc=np.sum)
```

```python
# 4 - 数据透视｜多方法
# 制作各省「销售总额」与「平均销售额」的数据透视表
df.pivot_table(index='省/自治区', values='销售额', aggfunc=[np.sum, np.mean])
```

```python
# 5 - 数据透视｜多指标
# 制作各省市「销售总额」与「利润总额」的数据透视表
df.pivot_table(values=['销售额', '利润'], index='类别', aggfunc=np.sum)
```

```python
# 6 - 数据透视｜多索引
# 制作「各省市」与「不同类别」产品「销售总额」的数据透视表
df.pivot_table(index=['省/自治区', '类别'], values= ['销售额'], aggfunc=np.sum)
```

```python
# 7 - 数据透视｜多层
# 制作各省市「不同类别」产品的「销售总额」透视表
pd.pivot_table(df,values = ['销售额'],index = ['省/自治区'], columns='类别',aggfunc = sum)
```

```python
# 8 - 数据透视｜综合
# 制作「各省市」、「不同类别」产品「销售量与销售额」的「均值与总和」的数据透视表，并在最后追加一行『合计』
pd.pivot_table(df, index=['省/自治区', '类别'], 
               values=['销售额', '数量'], 
               aggfunc=[np.mean, np.sum], 
               margins=True)
```

```python
# 9 - 数据透视｜筛选
# 在上一题的基础上，查询 「类别」 等于 「办公用品」 的详情
table = pd.pivot_table(df,values = ['销售额','数量'],index = ['省/自治区','类别'],aggfunc = ['mean',sum],margins=True)
table.query('类别 == ["办公用品"]')
```

```python
# 10 -数据透视｜逆透视
# 逆透视就是将宽的表转换为长的表，例如第五题的透视表进行逆透视，其中不要转换为列的数据是 数量 列

# melt(id_vars, value_vars, var_name, value_name, col_level, igonre_name)
table = pd.pivot_table(df,values = ['销售额','利润','数量'],index = '类别',aggfunc = sum)
print(table)
# table.melt(id_vars=['数量'],var_name='分类',value_name='金额')
table.melt(id_vars='数量', value_vars=['利润', '销售额'], value_name='金额', var_name='分类', ignore_index= True)
 
```

```python
# 数据合并
# cnocat - 数据拼接
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])


df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])


df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])


df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
```

```python
# 11 - concat｜默认拼接
# 拼接 df1 和 df2
pd.concat([df1, df2])
```

```python
# 12 - concat｜拼接多个
# 垂直拼接 df1、df2、df3，效果如下图所示
pd.concat([df1, df2, df3])
```

```python
# 13 - concat｜重置索引
# 垂直拼接 df1 和 df4，并按顺序重新生成索引
pd.concat([df1, df4], ignore_index=True)
```

```python
# 14 - concat｜横向拼接
# 横向拼接 df1、df4，效果如下图所示
pd.concat([df1, df4], axis=1)
```

```python
# 15 - concat｜横向拼接（取交集）
# 在上一题的基础上，只取结果的交集
pd.concat([df1, df4], axis=1, join='inner')
```

```python
# 16 - concat｜横向拼接（取指定）
# 在 14 题基础上，只取包含 df1 索引的部分
pd.concat([df1, df4], axis=1).reindex(df1.index)
```

```python
# 17 - concat｜新增索引
# 拼接 df1、df2、df3，同时新增一个索引（x、y、z）来区分不同的表数据来源
pd.concat([df1, df2, df3], keys=['x', 'y', 'z'])
```

```python
# merge - 数据连接
# concat是拼接， merge是连接。

# 18- merge | 按单键
# 根据 key 连接 left 和 right
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
pd.merge(left, right, on='key')
```

```python
# 19 - merge｜按多键
# 根据 key1 和 key2 连接 left 和 right
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})


right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})

left.merge(right, on=['key1', 'key2'])
```

```python
# 20 - merge｜左外连接
# 如下图所示的结果连接 left 和 right，保留左表全部键

left.merge(right, how='left', on=['key1', 'key2'])
```

```python
# 21 - merge｜右外连接
# 如下图所示的结果连接 left 和 right，保留右表全部键
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

```python
# 22 -merge｜全外连接
# 如下图所示的结果连接 left 和 right，保留全部键
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='cross')
```

```python
# 23 -merge｜内连接
# 如下图所示的结果连接 left 和 right，保留交集
pd.merge(left, right, how='inner', on=['key1', 'key2'])
```

```python
# 24 - merge｜重复索引
# 重新产生数据并按下图所示进行连接
left = pd.DataFrame({'k': ['K0', 'K1', 'K2'], 'v': [1, 2, 3]})
right = pd.DataFrame({'k': ['K0', 'K0', 'K3'], 'v': [4, 5, 6]})
pd.merge(left, right, on='k', suffixes=['_l', '_r'])
```

```python
# join - 组合
# 最后一个数据合并常用操作
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                     index=['K0', 'K1', 'K2'])


right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
# 25 - join｜左对齐
# 合并 left 和 right，并按照 left 的索引进行对齐
left.join(right)
```

```python
# 26 -join｜左对齐（外连接）
# 按下图所示进行连接

# 思考：merge 做法
left.join(right, how='outer')
```

```python
# 27 - join｜左对齐（内连接）
# 按下图所示进行连接
left.join(right, how='inner')
```

```python
# 28 -join｜按索引
# 重新产生数据并按下图所示进行连接（根据 key）
left = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'key': ['K0', 'K1', 'K0', 'K1']})


right = pd.DataFrame({'C': ['C0', 'C1'],
                      'D': ['D0', 'D1']},
                      index=['K0', 'K1'])
```

```python
# 29 - join｜按索引（多个）
# 重新产生数据并按下图所示进行连接（根据 key1 和 key2）
left.join(right, on='key')
```

```python
# 29 - join｜按索引（多个）
# 重新产生数据并按下图所示进行连接（根据 key1 和 key2）
left = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1']})


index = pd.MultiIndex.from_tuples([('K0', 'K0'), ('K1', 'K0'),
                                  ('K2', 'K0'), ('K2', 'K1')])


right = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                   'D': ['D0', 'D1', 'D2', 'D3']},
                  index=index)

print(left)
print(right)
left.join(right, on=['key1', 'key2'])
```
